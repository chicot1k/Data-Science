```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd

# Загрузка данных
data = pd.read_excel('/content/mg_no.xlsx')
X = data[['spectrs', 'mg']].values

# Рассчитаем инерцию для различных чисел кластеров
inertia = []
cluster_range = range(1, 11)
for n_clusters in cluster_range:
    kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
    inertia.append(kmeans.inertia_)

# Построим график правила локтя
plt.figure(figsize=(10, 6))
plt.plot(cluster_range, inertia, marker='o', linestyle='-')
plt.xlabel('Число кластеров')
plt.ylabel('Внутрикластерная сумма квадратов (Inertia)')
plt.title('Метод локтя для оптимального числа кластеров')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture

# Предположим, что у вас уже есть DataFrame df_f и данные X для кластеризации
X = df_f[['q1', 'q2', 'q3']]

# K-means кластеризация
kmeans = KMeans(n_clusters=3)
kmeans_labels = kmeans.fit_predict(X)

# GMM кластеризация
gmm = GaussianMixture(n_components=3)
gmm_labels = gmm.fit_predict(X)

# Визуализация результатов в 3D
fig = plt.figure(figsize=(12, 5))

# K-means результат
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(X['q1'], X['q2'], X['q3'], c=kmeans_labels, cmap='viridis')
ax1.set_title("K-means Clustering")
ax1.set_xlabel('Spectr')
ax1.set_ylabel('Q4')
ax1.set_zlabel('Q1')

# GMM результат
ax2 = fig.add_subplot(122, projection='3d')
ax2.scatter(X['q1'], X['q2'], X['q3'], c=gmm_labels, cmap='viridis')
ax2.set_title("GMM Clustering")
ax2.set_xlabel('Spectr')
ax2.set_ylabel('Q4')
ax2.set_zlabel('Q1')

plt.show()
```

```python
import plotly.express as px
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture

# Предположим, что у вас уже есть DataFrame df_f и данные X для кластеризации
X = df_f[['q1', 'q2', 'q3']]

# K-means кластеризация
kmeans = KMeans(n_clusters=3)
kmeans_labels = kmeans.fit_predict(X)

# GMM кластеризация
gmm = GaussianMixture(n_components=3)
gmm_labels = gmm.fit_predict(X)

# K-means график
fig_kmeans = px.scatter_3d(X, x='q1', y='q2', z='q3', color=kmeans_labels.astype(str), title="K-means Clustering")

# GMM график
fig_gmm = px.scatter_3d(X, x='q1', y='q2', z='q3', color=gmm_labels.astype(str), title="GMM Clustering")

# Отобразим графики
fig_kmeans.show()
fig_gmm.show()
```
